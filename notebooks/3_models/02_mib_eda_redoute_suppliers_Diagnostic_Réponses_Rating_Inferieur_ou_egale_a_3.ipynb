**Nettoyage de Texte**

```python
import string

# Définition de l'option d'affichage pour afficher toutes les lignes
#pd.set_option('display.max_colwidth', None)

# Fonction pour nettoyer le texte
def nettoyer_texte(texte):
    # Enlever la ponctuation
    texte_sans_ponctuation = texte.translate(str.maketrans('', '', string.punctuation))

    # Convertir le texte en minuscules
    texte_minuscules = texte_sans_ponctuation.lower()

    # Enlever les chiffres (si nécessaire)
    texte_sans_chiffres = ''.join([i for i in texte_minuscules if not i.isdigit()])

    return texte_sans_chiffres

# Appliquer la fonction de nettoyage à la colonne 'SupplierReply'
df_ratings_1_3['SupplierReply_Cleaned'] = df_ratings_1_3['SupplierReply'].apply(nettoyer_texte)

# Afficher les résultats
print(df_ratings_1_3.iloc[:10])
```

**2/ Identification du Personnel Répondant**

```python
import spacy

# Charger le modèle français
nlp = spacy.load('fr_core_news_sm')

# Fonction pour extraire les noms
def extraire_noms(texte):
    doc = nlp(texte)
    return [ent.text for ent in doc.ents if ent.label_ == 'PER']

# Appliquer la fonction aux réponses
df_ratings_1_3['Noms'] = df_ratings_1_3['SupplierReply'].apply(extraire_noms)
```

```python
df_ratings_1_3[['SupplierReply','Noms']].head(15)
```

Pour analyser plus en détail, il serait utile de séparer les réponses attribuées à "Ilham" de celles attribuées aux autres intervenants pour voir si Ilham répond différemment. De plus, calculer le nombre total de réponses par intervenant pourrait révéler si certains membres de l'équipe sont plus engagés dans les réponses personnalisées.

**Séparation des réponses d'Ilham et faire le cumul des réponses par intervenant.**

```python
# Séparer les réponses d'Ilham
df_ilham = df_ratings_1_3[df_ratings_1_3['Noms'].apply(lambda x: 'Ilham' in x)]

# Séparer les réponses des autres intervenants
df_autres = df_ratings_1_3[df_ratings_1_3['Noms'].apply(lambda x: 'Ilham' not in x)]

# Faire le cumul des réponses par intervenant
cumul_intervenants = df_ratings_1_3.explode('Noms')['Noms'].value_counts()

print("Réponses d'Ilham :")
print(df_ilham)
print("\nRéponses des autres intervenants :")
print(df_autres)
print("\nCumul des réponses par intervenant :")
print(cumul_intervenants)
```

Cette analyse révèle plusieurs points intéressants :

La présence de Noms Personnels : Les noms "Ilham", "Soukaina", "Yousra", "Loubna", et "Zineb" sont mentionnés plusieurs fois, indiquant que certaines réponses incluent des noms d'employés, ce qui pourrait être un indicateur de personnalisation des réponses.

Fréquence des Interventions : Ilham semble être l'intervenant le plus fréquent dans les réponses, suivie de près par "Wassim". Cela peut indiquer que ces employés sont les plus actifs ou les plus visibles dans les échanges avec les clients.

Variabilité de la Personnalisation : La présence de réponses sans aucun nom suggère une certaine variabilité dans la personnalisation des réponses. Certaines réponses semblent être standardisées (sans mention de nom), tandis que d'autres incluent des noms spécifiques.

Formules de Politesse : L'occurrence de formules telles que "Bonne journée" suggère l'utilisation de phrases de clôture standardisées.

L'analyse des noms suggère une tentative de personnalisation, mais la présence récurrente des mêmes noms pourrait également indiquer que ces mentions font partie d'un template de réponse.

**4/ Analyse de Sentiment**

```python
!pip install transformers
```

```python
from transformers import pipeline


# Charger le pipeline 'sentiment-analysis' avec le modèle multilingue
nlp_sentiment = pipeline('sentiment-analysis', model='nlptown/bert-base-multilingual-uncased-sentiment')

# Fonction pour obtenir le score de sentiment
def sentiment_score(texte):
    try:
        # Prendre uniquement le score du premier résultat
        result = nlp_sentiment(texte[:512])[0]  # Tronquer le texte à 512 tokens si nécessaire
        return {'label': result['label'], 'score': result['score']}
    except Exception as e:
        # En cas d'erreur, renvoyer un sentiment neutre par défaut
        return {'label': 'NEUTRAL', 'score': 0.0}

# Appliquer la fonction aux réponses
df_ratings_1_3['Sentiment'] = df_ratings_1_3['SupplierReply'].apply(sentiment_score)

```

**Analyse sur application BERT**

L'output de l'analyse de sentiment montre que les réponses du service client sont classées avec des étoiles, semblable à des avis produits ou services, allant de '1 star' à '5 stars'. Il est intéressant de remarquer que même pour des ratings clients faibles (comme 1 ou 2), le modèle a attribué des sentiments positifs ('5 stars') à certaines réponses. Cela pourrait indiquer que les réponses sont formulées de manière positive ou du moins neutre, ce qui est cohérent avec une tentative de maintenir un ton constructif et professionnel, même face à des retours négatifs.

Il y a cependant des réponses qui ont reçu un score de '1 star' en termes de sentiment, ce qui peut refléter un ton négatif ou une reconnaissance du mécontentement du client dans la réponse. Il est possible que dans ces cas, le modèle détecte des éléments de langage qui sont typiquement associés à des retours négatifs ou à des excuses.

Un autre point à souligner est la présence de réponses classées '3 stars', qui pourraient indiquer un ton neutre. Cela pourrait être dû à l'utilisation de formules standardisées qui ne transmettent ni un sentiment très positif ni négatif.

**Analyse croise sur coherence des rating de 1 a 5 avec l'analyse de sentiment BERT**

Majorité de Réponses Positives : La plupart des réponses ont reçu un score de sentiment de '5 stars', ce qui suggère que le ton des réponses est majoritairement positif.les réponses aux avis positifs sont formulées de manière à renforcer cette positivité.

Présence de Réponses avec Sentiment '1 star' : Il y a cependant des cas où les réponses, malgré un rating client de 4 étoiles, sont classées comme '1 star' par l'analyse de sentiment. Cela pourrait indiquer une discordance entre le ton perçu par le modèle et le contenu réel de la réponse.

Comparaison avec les Ratings Inférieurs : Comparé aux réponses pour les ratings de 1 à 3 étoiles, où nous avons également observé une variation dans les scores de sentiment, il semble que le service client s'efforce de maintenir un ton globalement positif ou neutre, quel que soit le rating du client.
