```python
# import customised plotting functions from plot_mtrics.py script
import pandas as pd
import os

from plot_metrics import plot_metrics_and_time, plot_radar_metrics, plot_radar_mult
from plot_metrics import plot_radar_some, barplot_metric_mult, lineplot_metrics, stack_barplot
```

## metrics data

```python
# find all files with names *_perf.csv in the working directry
files = [f for f in os.listdir() if f.endswith("_perf.csv")]
data = pd.DataFrame({})

# load each file into a dataFrame and concat all into a single dataFrame
for f in files:
    data = pd.concat([data, pd.read_csv(f, index_col=0)])

data = data.reset_index().drop("index", axis=1)
data
```

```python
model_list = ["log_reg_bsl", "lsvc_bsl",
              "Logistic_reg_cvz",
              "Logistic_reg_tfidf",
              "dense_nn",
              "GRU_0",
              "GRU_ru"
              ]

data_short = data[data["model"].isin(model_list)].sort_values("accuracy")

metric_selected = ["model", "accuracy", 
                   "f1_macro", "recall_macro", "precision_macro", 
                   "recall_0", "precision_0"]
data_short = data_short[["model", "accuracy", "f1_macro", "recall_macro", "precision_macro", "recall_0", "precision_0", "time_train", "time_predict"]]
```

```python
# radar chart of "log_reg_bsl", "lsvc_bsl","Logistic_reg_cvz", ...
# ... "Logistic_reg_tfidf", "dense_nn", "GRU_0","GRU_ru" models

# radar chart to compare the 7 models performance
df = data_short.iloc[:, :-2]
plot_radar_mult(df)
```

```python
# Line plot of metrics for our 7 short listed models
lineplot_metrics(df)
```

```python
# time_train and time_predict comparison
stack_barplot(data_short, "time_train", "time_predict")
```

```python
# barplot of metrics 
metric_selected = ["accuracy", 
                   "f1_macro", "recall_macro", "precision_macro", 
                   "recall_0", "precision_0"]

for met in metric_selected:
    barplot_metric_mult(df,met)
```
