```python


# Renommage  des colonnes
nouveau_noms_colonnes = {
    'typography_heading-xs__jSwUz': 'Initiales',
    'link_internal__7XN06 href': 'URL_Profil',
    'typography_heading-xxs__QKBS8': 'Nom',
    'typography_body-m__xgxZ_': 'Nombre_Avis',
    'typography_body-m__xgxZ_ 2': 'Pays',
    'star-rating_starRating__4rrcf src': 'URL_Étoile',
    'typography_body-m__xgxZ_ 3': 'Date_Avis',
    'link_internal__7XN06 href 2': 'URL_Avis',
    'typography_heading-s__f7029': 'Titre_Avis',
    'typography_body-l__KUYFJ': 'Contenu_Avis',
    'typography_body-m__xgxZ_ 4': 'Date_Expérience',
    'typography_body-m__xgxZ_ 5': 'Date_Expérience_Label',
    'typography_body-m__xgxZ_ 6': 'Utile',
    'typography_body-m__xgxZ_ 7': 'Partager',
    'typography_body-m__xgxZ_ 8': 'Autre'
}
df.rename(columns=nouveau_noms_colonnes, inplace=True)

# Suppression  des colonnes non pertinentes
colonnes_a_supprimer = ['URL_Étoile', 'Date_Expérience_Label', 'Partager', 'Autre']
df.drop(columns=colonnes_a_supprimer, inplace=True)

# Affichage les informations et les premières lignes du DataFrame nettoyé
df.info()
print(df.head(15))
```

```python
# 1. Analyse des Fréquences:
print("Nombre d'avis par pays:")
print(df['Pays'].value_counts())
print("\n")

print("Fréquence des valeurs dans la colonne 'Utile':")
print(df['Utile'].value_counts())
print("\n")

print("Distribution des nombres d'avis par utilisateur:")
print(df['Nombre_Avis'].value_counts())
print("\n")
```

```python
# 2. Groupements:

# Convertir la colonne 'Nombre_Avis' en numérique
df['Nombre_Avis'] = df['Nombre_Avis'].str.extract('(\d+)').astype(float)

# Calculer la moyenne du nombre d'avis par pays
moyenne_avis = df.groupby('Pays')['Nombre_Avis'].mean()
print("Moyenne du nombre d'avis par pays:")
print(moyenne_avis)

print("Moyenne du nombre d'avis par pays:")
print(df.groupby('Pays')['Nombre_Avis'].mean())
print("\n")

print("Nombre total d'avis 'Utile' par pays:")
print(df[df['Utile'] == "Utile"].groupby('Pays').size())
print("\n")

print("Jours avec le plus grand nombre d'avis:")
print(df.groupby('Date_Avis').size().sort_values(ascending=False))
print("\n")
```

```python
# 3. Tableaux Croisés:
print("Distribution des avis 'Utile' par pays:")
print(pd.crosstab(df['Pays'], df['Utile']))
print("\n")

print("Relation entre le pays de l'utilisateur et la date de l'avis:")
print(pd.crosstab(df['Pays'], df['Date_Avis']))
print("\n")

print("Relation entre le nombre d'avis et l'utilité de l'avis:")
print(pd.crosstab(df['Nombre_Avis'], df['Utile']))
print("\n")
```

```python
## Preprocessing
```

```python
# Suppression entrees incoherentes

# Supposons que votre dataframe s'appelle df
df = df[df['Pays'] != 'Actualisé le 14 juil. 2019']
```

```python
# Suppression des lignes avec des valeurs manquantes
df.dropna(inplace=True)


```

```python
# Suppression de la colonne "Partager" si valeur constante

if 'Partager' in df.columns:
    # Supprimer la colonne "Partager" si elle est constante
    if df['Partager'].nunique() == 1:
        df.drop(columns=['Partager'], inplace=True)

```

```python
# conversion des types de donnees
# Convertir la colonne "Date_Avis" en type date
df['Date_Avis'] = pd.to_datetime(df['Date_Avis'], errors='coerce')
```

```python
# Normalisation des noms de pays
#df['Pays'] = df['Pays'].replace({'FR': 'France'})
```

```python
# gestion des entrees double
df.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Histogramme du nombre d'avis par pays
plt.figure(figsize=(15, 7))
sns.countplot(data=df, y='Pays')
plt.title('Nombre d\'avis par pays')
plt.show();
```

```python
# Diagramme à barres montrant le nombre total d'avis "utiles" par pays
```

```python
df['Utile'] = pd.to_numeric(df['Utile'], errors='coerce')
# print(df['Utile'].dtype)

plt.figure(figsize=(10, 10))
sns.barplot(data=df, y='Pays', x='Utile', estimator=sum)
plt.title('Nombre total d\'avis "utiles" par pays')
plt.show();
```

```python
# Boxplot montrant la distribution du nombre d'avis par pays:

plt.figure(figsize=(15, 7))
sns.boxplot(data=df, y='Pays', x='Nombre_Avis')
plt.title('Distribution du nombre d\'avis par pays')
plt.show();
```

```python
# Heatmap
heatmap_data = df.pivot_table(values='Nombre_Avis', index='Pays', columns='Date_Avis', aggfunc='sum')
plt.figure(figsize=(20, 10))
sns.heatmap(heatmap_data, cmap='YlGnBu')
plt.title('Nombre d\'avis par date et par pays')
plt.show();
```

```python
# Scatter plot montrant la relation entre le nombre total d'avis et le nombre d'avis jugés "utiles"

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Nombre_Avis', y='Utile')
plt.title('Relation entre le nombre total d\'avis et le nombre d\'avis "utiles"')
plt.show();
```
